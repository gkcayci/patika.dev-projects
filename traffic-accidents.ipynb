```python
# Kütüphaneleri import edelim
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Python'da uyarıları kapatalım
import warnings 
warnings.filterwarnings("ignore")
```

```python
# Verileri içeri aktaralım
df = pd.read_csv('accidents_2005_to_2007.csv')
df2 = pd.read_csv('accidents_2009_to_2011.csv')
df3 = pd.read_csv('accidents_2012_to_2014.csv')
```

```python
# Üç veri seti de aynı değişkenlere sahip olduğu için veri setlerini birleştirelim.
data = pd.concat([df, df2,df3], axis=0)
data.head()
```

```python
# Veri setinindeki istatistiksel değerlere bakalım
data.describe().T
```

```python
# Veri setinde boş değerler var mı kontrol edelim
data.isnull().values.any()
```

```python
# Veri setindeki 0 olan değerlerin toplamaıa bakalım
data.eq(0).sum()
```

```python
# Veri setinden rastgele 5 değer çekelim
data.sample(5)
```

```python
# Veri setindeki unique değerlere bakalım
for i in data.columns:
    print("{},({}) : \n{}".format(i, len(data[i].unique()), data[i].unique()))
```

```python
# NaN değer içeren sütünları silelim
data.dropna(axis=1, how="all", inplace=True)
data.head(3)
```

```python
# Accident_Index veri setimizde olmasa da olur o yüzden bunu da kaldırıyoruz. Aynı şekilde onunla ilişkili olan LSOA_of_Accident_Location değerini de kaldırıyoryuz.
data.drop(["Accident_Index","LSOA_of_Accident_Location"], axis=1, inplace=True)
data.head(3)
```

```python
# Date değişkenini daha kolay inceleyebilmek için month,day,hour olarak parçalara ayıralım.

data["Date"] = pd.to_datetime(data["Date"])
data["Month"] = data["Date"].dt.month
data["Day"] = data["Date"].dt.day
data["Hour"] = pd.to_datetime(data["Time"]).dt.hour
data.head(3)
```

### Veriyi görselleştirelim

```python
# Histogram
data.hist(bins = 50, figsize = (30,15))
plt.show()
```

```python
# Çeşitli değişkennlere göre kazalara bakalım

plt.figure(figsize=(20,15))

plt.subplot(2,2,1)
data["Police_Force"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Police Force")

plt.subplot(2,2,2)
data["Accident_Severity"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Accident Severity")

plt.subplot(2,2,3)
data["Number_of_Vehicles"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Number of Vehicles")

plt.subplot(2,2,4)
data["Number_of_Casualties"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Number of Casualties")
```

Elde ettiğimiz grafiklere göre :

* En çok kaza 1 numaralı police force bölgesinde olmuştur. Bunu nedeni de nüfusun yoğun olduğu bir bölge olması veya işlek bir bölge olması olabilir.
* Accident severity'e göre en çok kaza 3.seviyedir
* Numbers of vehicles'a göre ise en çok kaza 2 araç arasında yaşanmıştır.
* Numbers of casualties'a göre ise en çok kaza geçiren sayısı 1'dir.

```python
# Yaralı sayısına (Number of Casualties) göre en tehlikeli kazaların  yaşandığı tarihlere bakalım.

date_casualities = data.groupby(['Date'])['Date','Number_of_Casualties'].sum()
worst_accident_days = date_casualities.sort_values(by='Number_of_Casualties',ascending=False)
w_20_days = worst_accident_days.head(20)
w_20_days=w_20_days.reset_index()


plt.figure(figsize=(16,4))
ax = sns.barplot(w_20_days['Date'], w_20_days['Number_of_Casualties'], alpha=0.8)
plt.title("The most dangerous days to travel",fontsize=25)
locs, labels = plt.xticks()
plt.setp(labels, rotation=80)
plt.ylabel('casualities recorded', fontsize=12)
plt.xlabel('day of accident', fontsize=12)

plt.show()
```

```python
# Zamana göre kazalara bakalım

data["Day"] = data["Day"].astype(str)
data["Day"] = data["Day"].map({"1": "Monday", "2": "Tuesday", "3": "Wednesday", "4": "Thursday", "5": "Friday", "6": "Saturday", "7":"Sunday"})

data["Month"] = data["Month"].astype(str)
data["Month"] = data["Month"].map({"1": "January", "2": "February", "3": "March", "4": "April", "5": "May", "6": "June", "7": "July", "8": "August", "9": "September", "10": "October", "11": "November", "12": "December"})

plt.figure(figsize=(20,15))

plt.subplot(2,2,1)
data["Year"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Year")

plt.subplot(2,2,2)
data["Month"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Month")

plt.subplot(2,2,3)
data["Day"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Day")

plt.subplot(2,2,4)
data["Hour"].value_counts().plot(kind="bar", color='#69b3a2')
plt.title("Hour")

plt.show()
```

Elde ettiğimiz verilere göre :
* Year grafiğine göre en fazla kaza 2005 yılında yaşanmıştır.
* Month grafiğine göre en fazla kaza eylül ve ekim aylarında yaşanmıştır.
* Day grafiğine göree en fazla kaza ilgili ayın 11.günüde yaşanmıştır.
* Hour grafiğine göre ise en fazla kaza 17 sularında yaşanmıştır.
